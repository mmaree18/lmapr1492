```python
#Marée Maxime 27361900
#Imports utiles
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import DiffractionPattern
```

```python
#Téléchargement du ficher CIF
with MPRester("rzFNAXwVskd5hQE3") as m:
    structure = m.get_structure_by_material_id("mp-961673")
```

# Analyse du spectre infrarouge du cristal TiFeTe
On crée un XRDCalculator qui va connaitre les informations du rayonnement $X$ incident $Cu K\alpha$ de longueur d’onde $\lambda=1.54060 Å$.

Ensuite, on applique ce rayonnement sur la structure de notre cristal via la fonction get_pattern() et on va récupérer une instance DiffractionPattern qui contient les résultats de la diffraction sur notre cristal TiFeTe.

```python
analyzestru = SpacegroupAnalyzer(structure)
conventionnal = analyzestru.get_conventional_standard_structure()
xrdcalc = XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
diffrapattern = xrdcalc.get_pattern(conventionnal,scaled=True, two_theta_range=(0, 90))
```

# Représentation du diffractogramme
Représentation pour des angles $2\theta$ allant de $0^°$ à $90^°$

```python
xrdcalc.show_plot(conventionnal)
```

# Premier Pic du Diffractogramme

```python
print("Le premier pic du diffractogramme est 2\u03B8 =",diffrapattern.x[0],"°")
print("L'intensité associé à ce pic est", diffrapattern.y[0], "(sur une échelle de 100)")
print("L'indice hkl associé à ce pic est (h,k,l) = ",diffrapattern.hkls[0] [0] ['hkl'], "et il a une multiplicité de :",diffrapattern.hkls[0] [0] ['multiplicity'])
print("La distance interréticulaire associée à ces indices hkl est", diffrapattern.d_hkls[0],"Å")
```

# Explication sur le choix des indices hkl associés aux pics

J'ai réalisé le diffractogramme sur la maille conventionnelle du matériau qui est une maille cubique face centrées (Cubique F).
Ce type de maille a une convention pour le choix des indices hkl associés aux pics, les indices hkl doivent respecter les relations :

h+k = 2n

h+l = 2o

k+l = 2p

Où n, o et p sont des nombres entiers.

On remarque que ces 3 relations sont bien respectées pour les 3 choix effectués ci dessus.
